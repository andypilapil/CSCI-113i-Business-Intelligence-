```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/andypilapil/csci-113-lab-1/main/food_coded.csv'
df = pd.read_csv(url)
```

First, we look for columns with null values and clean them. Then, we create a list of columns which have missing values followed by the number of missing values per column.

```python
cols_with_missing = [col for col in df.columns
                     if df[col].isnull().any()]
```

```python
for i in cols_with_missing:
    print(i,df[i].isnull().sum())
```

```python
df.shape
```

**We currently have 125 rows and 61**

Let us now take a closer look at the data and try to replace the missing values using fillna() function. We filter down the columns with missing values that can be replaced with values that is appropriate for them.

```python
df['calories_day'].fillna(1,inplace=True) # 1 - i dont know how many calories i should consume
df['comfort_food_reasons_coded'].fillna(9,inplace=True) # 9 – none
df['cuisine'].fillna(6,inplace=True) # 6 – other
df['employment'].fillna(4,inplace=True) # 4  - other
df['exercise'].fillna(5,inplace=True) # 5 – Never
df['type_sports'].fillna('Nothing',inplace=True)
```

After replacing some of the missing values with appropriate values, we remove the rows with the rest of the missing values that can't be replaced.

```python
for i in cols_with_missing:
    df = df[~df[i].isnull()]
```

```python
df.shape
```

**We now have 106 rows and 61 columns left**

**GPA Column**

The GPA column contains float values in string format along with missing values and erroneous values. To address this, let us create a new column by first removing the '.'(decimal point) from the values, in order to make use of the .isdigit() function. This .isdigit() function helps you to identify if all the values in the string are digits.

```python
df['GPA'].value_counts()
```

```python
df.dropna(subset=['GPA'],inplace=True)
df['GPA_new'] = df['GPA'].str.replace(".","")
df = df[~df['GPA_new'].str.isdigit() == False]
df['GPA'] = df['GPA'].astype(float)
df.drop('GPA_new',axis=1,inplace=True)
```
